```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

columns=['Sepal length','Sepal width','Petal length','Petal width','Class_lables']

df = pd.read_csv('iris.data', names=columns)
df.head()

#Describe the data 

df.describe()

# Visualize the whole dataset

sns.pairplot(df,hue='Class_lables') 
```

```python
#Seprating the features of data 
data=df.values
X= data [:,0:4]
Y= data [:,4]
```

```python
#split train the data

X_train,X_test,y_train,y_test=train_test_split(X,Y, test_size=0.2)
print(X_test)
```

```python
#support vector machine algo
model_svc=SVC()
model_svc.fit(X_train,y_train)
```

```python
#prediction of SVM
prediction1 = model_svc.predict(X_test)
#Calculate the accuracy
print(accuracy_score(y_test,prediction1)*100)
for i in range(len(prediction1)):
  print(y_test[i],prediction1[i])

```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
model_LR=LogisticRegression()
model_LR.fit(X_train,y_train)
```

```python
#prediction 2 of Logistic Regression
prediction2 = model_LR.predict(X_test)
#Calculate the accuracy
print(accuracy_score(y_test,prediction2)*100)
for i in range(len(prediction2)):
  print(y_test[i],prediction2[i])
```

```python
# Decisiontree
from sklearn.tree import DecisionTreeClassifier
model_DT= DecisionTreeClassifier()
model_DT.fit(X_train,y_train)
```

```python
#prediction 3 of Decision tree
prediction3=model_DT.predict(X_test)
#Calculate the accuracy
print(accuracy_score(y_test,prediction3)*100)
```

```python
# A detailed classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, prediction1))
print(classification_report(y_test, prediction2))
print(classification_report(y_test, prediction3))
```
